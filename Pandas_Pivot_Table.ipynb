## An index are necessary for creating a Simple pivot table. Let’s use the brand as our index.

The Numerical values are automatically filled while the non-numerical values are omitted. 
Each cell is the averages/mean of the car’s Price,iot,Mileage,year. The default aggfunc is mean.


```python
pd.pivot_table(UsaCdf,index='brand')
```

## Creating Multiple indexes and specify its value and aggfunc.

```python
pd.pivot_table(data=UsaCdf,index=['brand','color'],values=['mileage','price'],aggfunc='mean').head(10)
```

```python
pd.pivot_table(data=UsaCdf,index=['year','brand','color'],values=['mileage','price'],aggfunc='sum').head(20)
```

## Advanced filtering

```python
import numpy as np

pd_table=pd.pivot_table(data=UsaCdf,
                       index=['brand','color','year'],
                        values=['price','mileage'],
                        aggfunc={'price':[np.sum,np.mean],'mileage':np.mean})
```

```python
pd_table.query('color=="black"')
```
